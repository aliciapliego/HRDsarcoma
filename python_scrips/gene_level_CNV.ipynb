### Filter by sarcoma type

Filter by sarcoma type and do rowsums and divide by ciybands.

```python
anno2_merged2_t = anno2_merged2.T
anno2_merged2_t = anno2_merged2_t[anno2_merged2_t['Type'] == 'MPNST']
anno2_merged2_high = anno2_merged2_t[anno2_merged2_t['HRD34'] == 'High'].T.iloc[7:1207,]
high_number = len(anno2_merged2_high.columns)
anno2_merged2_low = anno2_merged2_t[anno2_merged2_t['HRD34'] == 'Low'].T.iloc[7:1207,]
low_number = len(anno2_merged2_low.columns)
```

```python
anno2_merged2_high_normalized = anno2_merged2_high.sum(axis=1)/high_number
anno2_merged2_high_normalized = anno2_merged2_high_normalized.reset_index().rename(columns = {0:'high'})
anno2_merged2_low_normalized  = anno2_merged2_low.sum(axis=1)/low_number
anno2_merged2_low_normalized  = anno2_merged2_low_normalized.reset_index().rename(columns = {0:'low'})
anno_normalized = anno2_merged2_low_normalized.merge(anno2_merged2_high_normalized, on = 'index')
anno_normalized['difference']  =  anno_normalized['high']- anno_normalized['low']
```
