```python
import pandas as pd
```

```python
treatment_drug = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/clinical_TCGA.tsv', sep = '\t')
```

```python
treatment_drug.head(3)
```

```python
TCGA_general = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/HRDness_TCGA_sarcoma_all.tsv', sep = '\t')
```

```python
RI = treatment_RI.merge(TCGA_general, right_on = 'Case_ID', left_on = 'bcr_patient_barcode').drop_duplicates(subset = 'bcr_patient_barcode')
```

```python
drug = TCGA_general.merge(treatment_drug, left_on = 'Case_ID', right_on = 'bcr_patient_barcode').drop_duplicates(subset = 'Case_ID')
```

```python
drug['treatment'] = 'Chemotherapy'
drug = drug[['Case_ID', 'HRDscore', 'treatment', 'measure_of_response']]
```

```python
RI['treatment'] = 'Radiotherapy'
RI = RI[['Case_ID', 'HRDscore', 'treatment', 'measure_of_response']]
```

```python
drugs_tcga = pd.concat([RI, drug], axis = 0)
```

```python
drugs_tcga['HRD34'] = 'High'
drugs_tcga
```

```python
drugs_tcga.loc[drugs_tcga['HRDscore'] <34, 'HRD34'] = 'Low'
```

```python
cin = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/cindex_results_sarc.tsv', sep = '\t')
cin = cin.set_index('Sample ID')
```

```python
cin_sum = cin.sum(axis = 1)
```

```python
cin_sum_df = pd.DataFrame(cin_sum)
cin_sum_df = cin_sum_df.reset_index()
```

```python
cin_sum_df['Sample ID'] = cin_sum_df['Sample ID'].str[:-4]
cin_sum_df = cin_sum_df.rename(columns = {0:'total_cin'})
```

```python
drugs_tcga
```

```python
cin_sum_df_tcga = cin_sum_df.merge(drugs_tcga, left_on = 'Sample ID', right_on = 'Case_ID').drop_duplicates(subset = 'Sample ID')
```

```python
cin_sum_df_tcga.to_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/cindex_results_sarc_treatment_hrd.tsv', sep = '\t', index = None)
```
