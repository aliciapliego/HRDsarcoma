```python
import pandas as pd
```

```python
df_patients= pd.read_excel("SARC_Clinical_Data_Jan-25-2016_Freeze_Cases_Only (1).xlsx", sheet_name="SARC Patient")
```

```python
df_all = pd.read_csv("/Users/aliciapliego/Projects/HRD_project/allele_specific_tcga/annotations/sarcoma_all_annotated.tsv", sep = '\t')
```

```python
df_all['HRDness'] = df_all['LOH'] + df_all['LST'] + df_all['TAI']
```

```python
df_all = df_all[['Case ID', 'Tumor_type', 'LOH', 'LST', 'TAI']] 
```

```python
df_all.to_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/metadata_TCGA_general.tsv', sep = '\t', index = None)
```

```python
df_all_merged = df_all.merge(df_patients, left_on = 'Case ID', right_on = 'bcr_patient_barcode')
df_all_merged= df_all_merged[['Case ID', 'Tumor_type', 'LOH', 'LST', 'TAI', 'radiation_treatment_adjuvant','pharmaceutical_tx_adjuvant']] 
```

```python
df_drugs= pd.read_excel("SARC_Clinical_Data_Jan-25-2016_Freeze_Cases_Only (1).xlsx", sheet_name="SARC Drug")
```

```python
df_all_merged_drugs = df_all_merged.merge(df_drugs, left_on = 'Case ID', right_on = 'bcr_patient_barcode')
```

```python
df_all_merged_drugs = df_all_merged_drugs[['Case ID', 'Tumor_type', 'pharmaceutical_therapy_drug_name','treatment_best_response']]
```

```python
df_radiation = pd.read_excel("SARC_Clinical_Data_Jan-25-2016_Freeze_Cases_Only (1).xlsx", sheet_name="SARC Radiation")
```

```python
df_radiation
```

```python
df_all_merged_drugs_radi = df_all_merged_drugs.merge(df_radiation, left_on = 'Case ID', right_on = 'bcr_patient_barcode')
```

```python
df_all_merged_drugs_radi
```

```python
df_all_merged_drugs_radi = df_all_merged_drugs_radi[['Case ID', 'Tumor_type', 'pharmaceutical_therapy_drug_name','treatment_best_response_x', 'radiation_therapy_site', 'radiation_total_dose', 'treatment_best_response_y']]
```

```python
df_all_merged_drugs_radi = df_all_merged_drugs_radi.rename(columns = {'treatment_best_response_x':'treatment_best_response_chemo', 'treatment_best_response_y':'treatment_best_response_radiation'})
df_all_merged_drugs_radi.to_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/metadata_drug_treatment_TCGA.tsv', sep = '\t', index = None)
```
