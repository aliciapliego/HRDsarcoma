```python
sample_id = []
df_index = []
for file in files_genelevel:
    df = pd.read_csv(file, sep = '\t')
    df = df[df['chromosome'] != 'chrX']
    df = df[df['chromosome'] != 'chrY']
    sample_name = os.path.split(file)[-1]
    sample_id.append(sample_name)
    amp = []
    dele = []
    neu = []
    sample_id = []
    gene_name = []
    for index, row in df.iterrows():
        if row[5] > 2.25:
            value_amp = row[7] -2
            amp.append(value_amp)
            value_del = 0
            dele.append(value_del)
            value_neu = 0
            neu.append(value_neu)
            gene_name.append(row[1])

        elif row[5] < 1.75:
            value = 2 - row[7]
            dele.append(value)

            value_amp = 0
            amp.append(value_amp)
            value_neu = 0
            neu.append(value_neu)
            gene_name.append(row[1])

        else:
            sample_id.append(row[6])
            value = 2
            neu.append(value)
            value_amp = 0
            amp.append(value_amp)
            value_del = 0
            dele.append(value_del)
            gene_name.append(row[1])
            
    dict = {'gene_name': gene_name,'Amp': amp, 'Del': dele} 
    
    df = pd.DataFrame(dict)
    df['cin'] = df['Amp'] + df['Del']
    
    df = df[['gene_name','cin']]
    df = df.set_index('gene_name')
    df = df.rename(columns = {'cin':sample_name})
    df_index.append(df)

df_cin = pd.concat(df_index, axis=1)
```
