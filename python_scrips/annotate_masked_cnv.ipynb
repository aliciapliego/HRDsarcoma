##  Annotate masked_cnv data from TCGA 

```python
import pandas as pd
```

```python
cinmetrics_sarc = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/cinmetrics_sarcoma.tsv', sep = '\t')
```

```python
cinmetrics_sarc['sample_id_v2'] = cinmetrics_sarc['sample_id'].str[:-12]
cinmetrics_sarc['case_id'] = cinmetrics_sarc['sample_id'].str[:-16]
```

```python
cinmetrics_sarc
```

```python
samplesheet_sarc = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/intermediate_annotation_files/sample_sheet_masked_cnv_sarc.tsv', sep = '\t')
```

```python
samplesheet_sarc
```

```python
cinmetrics_sarc_anno = samplesheet_sarc.merge(cinmetrics_sarc, left_on = 'Sample ID', right_on = 'sample_id_v2')
```

```python
cinmetrics_sarc_anno =cinmetrics_sarc_anno[['sample_id','sample_id_v2', 'case_id','tai','cna','base_segments','break_points','fga', 'Sample Type']]
```

```python
cinmetrics_sarc_anno
```

```python
anno_tcga_sarc = pd.read_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/metadata_TCGA_general.tsv', sep = '\t')
```

```python
anno_tcga_sarc_tumortype = anno_tcga_sarc.merge(cinmetrics_sarc_anno, left_on = 'Case ID', right_on = 'case_id')
```

```python
anno_tcga_sarc_tumortype
```

```python
anno_tcga_sarc_tumortype_corr = anno_tcga_sarc_tumortype[['sample_id', 'Tumor_type','HRDscore', 'tai','cna','base_segments','break_points','fga','Sample Type', "HRD34"]]
```

```python
anno_tcga_sarc_tumortype_corr.to_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/CINmetrics_annotated_sarc_correlation_hrd.tsv',sep = '\t', index = None)
```

```python
anno_tcga_sarc_tumortype = anno_tcga_sarc_tumortype[['sample_id', 'Tumor_type', 'tai','cna','base_segments','break_points','fga','Sample Type', "HRD34"]]
```

```python
anno_tcga_sarc_tumortype.loc[anno_tcga_sarc_tumortype['Sample Type'] == "Blood Derived Normal", 'HRD34'] = 'None'
```

```python
anno_tcga_sarc_tumortype.loc[anno_tcga_sarc_tumortype['Sample Type'] == "Solid Tissue Normal", 'HRD34'] = 'None'
```

```python
anno_tcga_sarc_tumortype.to_csv('/Users/aliciapliego/Projects/HRDsarcoma/data/CINmetrics_annotated_sarc.tsv',sep = '\t', index = None)
```
